```python
import nltk
from wordcloud import WordCloud
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import string

# Load your dataset
df = pd.read_csv('spamclass.csv')

# Pre-processing functions
def clean_text(text):
    # Remove punctuation
    text = text.translate(str.maketrans("", "", string.punctuation))
    # Convert to lowercase
    text = text.lower()
    # Remove stopwords
    stop_words = set(stopwords.words("english"))
    text_tokens = text.split()
    text = [word for word in text_tokens if word not in stop_words]
    # Lemmatize words
    lemmatizer = WordNetLemmatizer()
    text = [lemmatizer.lemmatize(word) for word in text]
    # Rejoin words
    text = ' '.join(text)
    return text

def preprocess_data(df):
    # Remove NaN values
    df = df.dropna()
    # Separate spam and ham messages
    spam_messages = df[df['Category'] == 'spam']['Message'].tolist()
    ham_messages = df[df['Category'] == 'ham']['Message'].tolist()
    # Clean messages
    spam_messages = [clean_text(message) for message in spam_messages]
    ham_messages = [clean_text(message) for message in ham_messages]
    return spam_messages, ham_messages

# Preprocess data
spam_messages, ham_messages = preprocess_data(df)

# Create word clouds
spam_text = ' '.join(spam_messages)
ham_text = ' '.join(ham_messages)

spam_wc = WordCloud(width=600, height=400, background_color='white').generate(spam_text)
ham_wc = WordCloud(width=600, height=400, background_color='white').generate(ham_text)

# Plot word clouds
plt.figure(figsize=(10, 8))
plt.subplot(1, 2, 1)
plt.imshow(spam_wc, interpolation='bilinear')
plt.title('Word Cloud for Spam Messages')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(ham_wc, interpolation='bilinear')
plt.title('Word Cloud for Ham Messages')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import re
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import LabelEncoder

# Assuming you have a DataFrame 'df' with columns 'Category' and 'Messages'
# Replace 'your-dataset.csv' with the actual file path and name
# Load your dataset
df = pd.read_csv('spamclass.csv')

# Remove rows with NaN values in the 'Message' column
df = df.dropna(subset=['Message'])

# Convert the 'Category' column to numerical values
le = LabelEncoder()
df['Category'] = le.fit_transform(df['Category'])

# Extract numerical features from the 'Message' column
df['NumWords'] = df['Message'].apply(lambda x: len(str(x).split()))
df['NumChars'] = df['Message'].apply(lambda x: len(str(x)))
df['NumUppercase'] = df['Message'].apply(lambda x: len(re.findall('[A-Z]', str(x))))

# Create a pairplot
sns.pairplot(df, hue='Category', vars=['NumWords', 'NumChars', 'NumUppercase'])

# Show the plot
plt.show()
```

```python
                                                         # Data Preprocessing
# Importing the Porter Stemmer for text stemming
from nltk.stem.porter import PorterStemmer

# Importing the string module for handling special characters
import string

# Creating an instance of the Porter Stemmer
ps = PorterStemmer()

# Lowercase transformation and text preprocessing function
def transform_text(text):
    # Transform the text to lowercase
    text = text.lower()
    
    # Tokenization using NLTK
    text = nltk.word_tokenize(text)
    # Removing special characters
    y = []
    for i in text:
        if i.isalnum():
            y.append(i)
            
    # Removing stop words and punctuation
    text = y[:]
    y.clear()
    
    # Loop through the tokens and remove stopwords and punctuation
    for i in text:
        if i not in stopwords.words('english') and i not in string.punctuation:
            y.append(i)
            # Stemming using Porter Stemmer
    text = y[:]
    y.clear()
    for i in text:
        y.append(ps.stem(i))
    
    # Join the processed tokens back into a single string
    return " ".join(y)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Assuming you have a DataFrame 'df' with columns 'Category' and 'Messages'
# Replace 'your-dataset.csv' with the actual file path and name
# Load your dataset
df = pd.read_csv('spamclass.csv')

# Remove rows with NaN values in the 'Message' column
df = df.dropna(subset=['Message'])

# Filter DataFrame to include only rows with 'Category' labeled as spam
spam_df = df[df['Category'] == 'spam']

# Define a list of stopwords
stopwords_list = ['to', 'you', 'and', 'the', 'is', ...]  # Add more stopwords as needed

# Extract the words from the spam messages and remove stopwords
vectorizer = CountVectorizer(stop_words=stopwords_list)
all_words_spam = vectorizer.fit_transform(spam_df['Message'])

# Count the frequencies of the words
word_freqs_spam = pd.DataFrame(all_words_spam.sum(axis=0).tolist()[0], columns=['Frequency'])
word_freqs_spam['Word'] = vectorizer.get_feature_names_out()

# Filter the DataFrame to only include words with a frequency greater than 100
word_freqs_spam = word_freqs_spam[word_freqs_spam['Frequency'] > 100]

# Sort the DataFrame by frequency
word_freqs_spam = word_freqs_spam.sort_values(by='Frequency', ascending=False)

# Create a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(data=word_freqs_spam, x='Word', y='Frequency', palette='bright')

# Rotate the x-axis labels
plt.xticks(rotation=90)

# Set plot title and labels
plt.title('Most Frequent Words in Spam Messages')
plt.xlabel('Word')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```
